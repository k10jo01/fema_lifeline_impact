# Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./final_df.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.head()
```

```python
df['lifeline'].value_counts().sort_index()
```

```python
# bar chart:

# (figure size)
plt.figure(figsize=(10,6))

# passing in X (labels) and Y (values), could add colors?
plt.bar(df['lifeline'].value_counts().index.tolist(), 
        df['lifeline'].value_counts())

# Add a title and show the plot.
plt.title('Number of establishments for each FEMA Lifeline', fontsize=20)
plt.xlabel('FEMA Lifeline Number', fontsize = 15)
plt.ylabel('Number of establishments', fontsize = 15)
plt.xticks(size=20)
plt.yticks(size=15);
```

```python
# using seaborn:
plt.figure(figsize=(10,6))
sns.countplot(data=df, x = 'lifeline')
plt.title('Number of establishments for each FEMA Lifeline', fontsize=20)
plt.xlabel('FEMA Lifeline Number', fontsize = 15)
plt.ylabel('Number of establishments', fontsize = 15)
plt.xticks(size=15)
plt.yticks(size=15);
```
